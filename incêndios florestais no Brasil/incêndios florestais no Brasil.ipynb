```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.io as pio
import plotly.express as px
import plotly.io as pio

# Substitua 'YOUR_MAPBOX_ACCESS_TOKEN' pela sua chave de API do Mapbox
#pio.kaleido.scope.default_mapbox_access_token = 'YOUR_MAPBOX_ACCESS_TOKEN'


# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
# Lendo os dados
Base_Dados = pd.read_csv(r'C:\Users\rafae\OneDrive\Área de Trabalho\Port entrevista\Python analise de dados\incêndios florestais no Brasil\Dados_Indendio.csv', encoding='latin-1')

# Verificando
Base_Dados.head()
```

```python
# Nulos Plot
plt.figure( figsize=(14,5) )
plt.title('Análise de campos nulos')
sns.heatmap( Base_Dados.isnull(), cbar=False );
```

```python
# analise por ano dos incendios
Analise = Base_Dados.groupby( by=['year'] ).sum().reset_index()

Analise.head()

# Tamanho
plt.figure( figsize=(12, 5) )

# Style
plt.style.use('ggplot')

# Grafico
plt.title( 'Total indêncidios no Brasil: 1997 - 2017', loc='left', fontsize=14 )
sns.lineplot( data=Analise, x='year', y='number', estimator='sum', lw=2, color='#ff5555', alpha=0.85 );

# Labels
plt.xlabel('Quantidade')
plt.ylabel('Período');
```

```python
# analise por ano dos incendios
Analise_02 = Base_Dados.groupby( by=['year', 'month'] ).sum().reset_index()
Analise_02.head()

# Tamanho
plt.figure( figsize=(12, 5) )

# Grafico
plt.title( 'Indêncidios por mês', loc='left', fontsize=14 )
sns.boxplot( data=Analise_02, x='month', y='number', palette='coolwarm', saturation=1, width=0.9, linewidth=2,
            order=['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'] )

# Labels
plt.xlabel('Mês')
plt.ylabel('Número de incêndios');
```

```python
# analise por ano dos incendios
Analise_03 = Base_Dados.groupby( by=['state'] ).sum().reset_index()[['state', 'number']].sort_values( 'number', ascending=False )
Analise_03.head()

# Tamanho
plt.figure( figsize=(12, 5) )

# Grafico
plt.title( 'Estados com maior número de incêndios', loc='left', fontsize=14 )

# Grafico
plt.bar( Analise_03.state, Analise_03['number'], color='#f44e3f' )

# Labels
plt.ylabel('Quantidade')
plt.xticks( rotation=90 );
```

```python
# Estados TOP 10
Lista_TOP10 = Analise_03['state'][0:10].values

# Tamanho
plt.figure( figsize=(12, 5) )

# Grafico
plt.title( 'Top 10 Estados com incêncios', loc='left', fontsize=14 )

# Loop
for Coluna in Lista_TOP10:
  
  # Filtrar o estado
  Filtro = Base_Dados.loc[ Base_Dados['state'] == Coluna ]

  # Agrupar os valores para sumarizar
  Analise_Local = Filtro.groupby( by=['year'] ).sum().reset_index()

  # Plot
  sns.lineplot(data=Analise_Local, x='year', y='number', lw=2, alpha=0.85)

# Labels
plt.xlabel('Período')
plt.ylabel('Número de incêndios')

# Legenda
plt.legend( Lista_TOP10, bbox_to_anchor=(1, 0.7) );
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Exemplo de carregamento dos dados (substitua pelo seu DataFrame real)
# Base_Dados = pd.read_csv('caminho_para_seus_dados.csv')

# 1. Análise por ano dos incêndios
Analise = Base_Dados.groupby(by=['year']).sum().reset_index()

# Gráfico Total de Incêndios no Brasil: 1997 - 2017
plt.figure(figsize=(12, 5))
plt.style.use('ggplot')
plt.title('Total de Incêndios no Brasil: 1997 - 2017', loc='left', fontsize=14)
sns.lineplot(data=Analise, x='year', y='number', estimator='sum', lw=2, color='#ff5555', alpha=0.85)
plt.xlabel('Ano')
plt.ylabel('Quantidade de Incêndios')
plt.grid()
plt.show()

# 2. Análise por mês dos incêndios
Analise_02 = Base_Dados.groupby(by=['year', 'month']).sum().reset_index()

# Gráfico Incêndios por Mês
plt.figure(figsize=(12, 5))
plt.title('Incêndios por Mês', loc='left', fontsize=14)
sns.boxplot(data=Analise_02, x='month', y='number', palette='coolwarm', saturation=1, width=0.9, linewidth=2,
            order=['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'])
plt.xlabel('Mês')
plt.ylabel('Número de Incêndios')
plt.grid()
plt.show()

# 3. Análise por estado
Analise_03 = Base_Dados.groupby(by=['state']).sum().reset_index()[['state', 'number']].sort_values('number', ascending=False)

# Gráfico Estados com Maior Número de Incêndios
plt.figure(figsize=(12, 5))
plt.title('Estados com Maior Número de Incêndios', loc='left', fontsize=14)
plt.bar(Analise_03.state, Analise_03['number'], color='#f44e3f')
plt.ylabel('Quantidade de Incêndios')
plt.xticks(rotation=90)
plt.grid()
plt.show()

# 4. Top 10 Estados com Incêndios
Lista_TOP10 = Analise_03['state'][0:10].values

plt.figure(figsize=(12, 5))
plt.title('Top 10 Estados com Incêndios', loc='left', fontsize=14)
for Coluna in Lista_TOP10:
    Filtro = Base_Dados.loc[Base_Dados['state'] == Coluna]
    Analise_Local = Filtro.groupby(by=['year']).sum().reset_index()
    sns.lineplot(data=Analise_Local, x='year', y='number', lw=2, alpha=0.85)
plt.xlabel('Ano')
plt.ylabel('Número de Incêndios')
plt.legend(Lista_TOP10, bbox_to_anchor=(1, 0.7))
plt.grid()
plt.show()

# 5. Mapa de Calor Geográfico
# Certifique-se de que o DataFrame 'Analise_Geografica' contém as colunas 'Latitude', 'Longitude' e 'Incêndios'
# Analise_Geografica = Base_Dados[['Latitude', 'Longitude', 'Incêndios']]
# Exemplo: Analise_Geografica = pd.DataFrame({'Latitude': [lat1, lat2], 'Longitude': [lon1, lon2], 'Incêndios': [value1, value2]})

# Fazendo o plot
mapbox_style = 'open-street-map'
fig = px.density_mapbox(
    Analise_Geografica,
    lat='Latitude',
    lon='Longitude',
    z='Incêndios',
    radius=30,
    center=dict(lat=-12.700, lon=-46.5555),
    zoom=3,
    mapbox_style=mapbox_style
)
fig.update_layout(title='Distribuição Geográfica de Incêndios')
fig.show()

# Conclusão
conclusao = """
A análise dos dados sobre incêndios florestais no Brasil entre 1997 e 2017 revelou padrões significativos. 
Observou-se uma variação temporal e espacial na ocorrência de incêndios, indicando que certas regiões e períodos do ano são mais vulneráveis. 
Os resultados destacam a necessidade de intervenções específicas em meses e estados críticos, 
além de sugerir a implementação de políticas de monitoramento e educação ambiental.

"""

print(conclusao)

```
